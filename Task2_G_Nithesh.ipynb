```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your cleaned dataset
df = pd.read_csv("cleaned_titanic (1).csv")

#summary statictics
df.info()
df.describe()


# Histogram + Boxplot
fig, axes = plt.subplots(2, 1, figsize=(10, 6), sharex=True)

sns.histplot(df['Age'].dropna(), ax=axes[0], color='skyblue', kde=True)
axes[0].set_title('Age Distribution - Histogram')

sns.boxplot(x=df['Fare'], ax=axes[1], color='lightgreen')
axes[1].set_title('Fare Distribution - Boxplot')

plt.tight_layout()
plt.show()
plt.figure(figsize=(6, 4))

# countplot
sns.countplot(
    data=df,
    x='Survived',
    hue='Sex',
    palette=sns.color_palette(['steelblue', 'salmon'])  # NOT using dict, use a list instead
)

plt.title('Survival Count by Gender')
plt.show()

# Correlation Heatmap
plt.figure(figsize=(8, 6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# Pairplot
selected_cols = ['Age', 'Fare', 'Pclass', 'Survived']
pairplot_df = df[selected_cols].dropna()

sns.pairplot(pairplot_df, hue='Survived', palette='husl')
plt.suptitle("Pairplot of Key Features", y=1.02)
plt.show()

#4. Inference and Patterns
    #Most passengers who survived were women.
    #Age has a slight skew; many passengers were young.
    #Fare has extreme outliers.
    #Strong correlation between Pclass and Fare.
```
